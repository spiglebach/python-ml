```python
import cv2.cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_img():
    blank_img = np.zeros((600,600))
    font = cv.FONT_HERSHEY_SIMPLEX
    cv.putText(blank_img,text='ABCDE',org=(50,300),fontFace=font,fontScale=5,color=(255,255,255),thickness=25)
    return blank_img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(8,6))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img = load_img()
display_img(img)
```

```python
kernel = np.ones((5,5),dtype=np.uint8)
kernel
```

```python
result = cv.erode(img,kernel,iterations=1)
display_img(result)
```

```python
result = cv.erode(img,kernel,iterations=4)
display_img(result)
```

```python
img = load_img()
white_noise = np.random.randint(low=0,high=2,size=(600,600))
display_img(white_noise)
```

```python
white_noise = white_noise * 255
noise_img = white_noise + img
display_img(noise_img)
```

```python
opening = cv.morphologyEx(noise_img,cv.MORPH_OPEN,kernel)
display_img(opening)
```

```python
img = load_img()
black_noise = np.random.randint(low=0,high=2,size=(600,600))
black_noise = black_noise * -255
black_noise_img = img + black_noise
black_noise_img[black_noise_img == -255] = 0
display_img(black_noise_img)
```

```python
closing = cv.morphologyEx(black_noise_img,cv.MORPH_CLOSE,kernel)
display_img(closing)
```

```python
img = load_img()
display_img(img)
```

```python
gradient = cv.morphologyEx(img,cv.MORPH_GRADIENT,kernel)
display_img(gradient)
```
