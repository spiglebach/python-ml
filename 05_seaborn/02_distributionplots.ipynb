```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../course_notebooks/05-Seaborn/dm_office_sales.csv')
df.head()
```

```python
sns.rugplot(df,x='salary',height=.5)
```

```python
sns.set(style='darkgrid')
sns.displot(df,x='salary',bins=10,color='red',edgecolor='yellow',kde=True,rug=True)
```

```python
sns.histplot(df,x='salary')
```

```python
sns.kdeplot(data=df,x='salary')
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0,100,200)
sample_ages = pd.DataFrame(sample_ages,columns=['age'])
sample_ages
```

```python
sns.rugplot(data=sample_ages,x='age')
sns.displot(data=sample_ages,x='age',bins=30,kde=True)
```

```python
sns.kdeplot(sample_ages,x='age',clip=[0,100],bw_adjust=0.4,fill=True)
```
