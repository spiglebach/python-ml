```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../course_notebooks/05-Seaborn/StudentsPerformance.csv')
df.head()
```

```python
sns.jointplot(data=df,x='math score',y='reading score')
```

```python
sns.jointplot(data=df,x='math score',y='reading score',kind='hex')
```

```python
sns.jointplot(data=df,x='math score',y='reading score',kind='hist')
```

```python
sns.jointplot(data=df,x='math score',y='reading score',kind='kde',fill=True)
```

```python
sns.jointplot(data=df,x='math score',y='reading score',hue='gender',alpha=.6)
```

```python
sns.pairplot(df,hue='gender',corner=True)
```
