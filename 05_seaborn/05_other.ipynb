```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../course_notebooks/05-Seaborn/StudentsPerformance.csv')
df.head()
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='box',col='lunch',row='test preparation course')
```

```python
g=sns.PairGrid(df,hue='gender')
g=g.map_upper(sns.scatterplot)
g=g.map_diag(sns.kdeplot,fill=True)
g=g.map_lower(sns.kdeplot)

g=g.add_legend()
```
