```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../course_notebooks/05-Seaborn/dm_office_sales.csv')
df.head()
```

```python
plt.figure(figsize=(12,4))
sns.scatterplot(x='salary',y='sales',data=df)
```

```python
plt.figure(figsize=(12,4))
sns.scatterplot(x='salary',y='sales',data=df,hue='level of education')
```

```python
plt.figure(figsize=(12,4))
sns.scatterplot(x='salary',y='sales',data=df,hue='salary')
```

```python
plt.figure(figsize=(12,4))
sns.scatterplot(x='salary',y='sales',data=df,hue='level of education',palette='Dark2')
```

```python
plt.figure(figsize=(12,4))
sns.scatterplot(x='salary',y='sales',data=df,hue='level of education',size='work experience',alpha=.7,style='training level')
plt.savefig('myplot.jpg',bbox_inches='tight')
```
