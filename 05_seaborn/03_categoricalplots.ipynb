```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../course_notebooks/05-Seaborn/dm_office_sales.csv')
df.head()
```

```python
plt.figure(figsize=(10,4))
sns.countplot(data=df,x='division')
```

```python
plt.figure(figsize=(10,4))
sns.countplot(data=df,x='level of education', hue='division',palette='Set3')
```

```python
plt.figure(figsize=(10,4))
sns.barplot(data=df,x='level of education',y='salary',estimator=np.mean,errorbar='sd',hue='division')
plt.legend(bbox_to_anchor=(1.05,1))
```

```python
ds = pd.read_csv('../course_notebooks/05-Seaborn/StudentsPerformance.csv')
ds.head()
```
