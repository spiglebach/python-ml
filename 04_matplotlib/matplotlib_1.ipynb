```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
plt.plot(x,y);
```

```python
plt.plot(x,y,'bo')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.xlim(0,10)
plt.ylim(0,20)
plt.title('Stuff')
plt.savefig('myfirstplot.jpg')
```

```python
a=np.linspace(0,10,11)
b=a**4
f = plt.figure(figsize=(10,3),dpi=100)
a1 = f.add_axes([0,0,.5,1])
a1.set_title('Yoooo')
a1.plot(a,b)

az = f.add_axes([.6,0,.5,.4])
az.set_title('zoomed')
az.set_xlim(4,6)
az.set_ylim(200,1500)
az.plot(a,b)

a2 = f.add_axes([0.6,0.6,.5,.4])
a2.set_title('srsly')
a2.plot(x,y)

f.savefig('fig.jpg',bbox_inches='tight')
```

```python
a=np.linspace(0,10,11)
b=a**4
x = np.arange(0,10)
y = 2*x
fig, axes = plt.subplots(2,2)
axes[0,0].plot(a,b,label='X vs X^4',color='red',ls='--')
axes[0,0].plot(a,a**3.5,label='X vs X^3.5',color=(0,1,.5))
lines = axes[0,0].plot(a,b+2000,label='X vs X^4+2000',color=(0,.2,.8),lw=4)
lines[0].set_dashes([1,2,1,2,10,2])

axes[0,0].legend(loc=(0,1.1))

axes[0,1].plot(x,y)
axes[0,1].plot(x,y,marker='o',markerfacecolor='red',markeredgewidth=3)
axes[1,0].plot(x,y,marker='+',ms=20)
axes[1,1].plot(x,y)
fig.suptitle('Figure level title')
fig.subplots_adjust(wspace=.3,hspace=.4)

fig.savefig('subplots.jpg',bbox_inches='tight')
```
