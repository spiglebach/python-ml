#### TASK: Open the *dog_backpack.jpg* image from the DATA folder and display it in the notebook. Make sure to correct for the RGB order.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
img = cv2.imread('../DATA/dog_backpack.jpg')
plt.imshow(img)
```

```python
img.shape
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
plt.imshow(cv2.flip(img,0))
```

```python
cv2.rectangle(img,(200,380),(610,720),color=(0,255,0),thickness=5)
plt.imshow(img)
```

```python
pts = np.array([[400,400],[300,650],[500,650]])
img2 = img.copy()
cv2.polylines(img2,[pts],isClosed=True,color=(0,0,255),thickness=10)
plt.imshow(img2)
```

```python
img3 = img.copy()
cv2.fillPoly(img3,[pts],color=(255,255,0))
plt.imshow(img3)
```
