```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('../course_notebooks/DATA/cluster_mpg.csv')
df.head()
```

```python
df.describe()
```

```python
df['origin'].value_counts()
```

```python
df_with_dummies = pd.get_dummies(df.drop('name',axis=1))
df_with_dummies
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df_with_dummies)
scaled_data
```

```python
scaled_df = pd.DataFrame(scaled_data,columns=df_with_dummies.columns)
scaled_df
```

```python
sns.heatmap(scaled_df)
```

```python
sns.clustermap(scaled_df.drop(['origin_europe','origin_japan','origin_usa'],axis=1),row_cluster=False)
```

```python
sns.clustermap(scaled_df,col_cluster=False)
```

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters=4)
cluster_labels = model.fit_predict(scaled_df)
cluster_labels
```

```python
sns.scatterplot(data=df,x='mpg',y='horsepower',hue=cluster_labels,palette='viridis')
```

```python
len(scaled_df.columns)
```

```python
np.sqrt(10) # Max distance possible between two points (min/max scaling)
```

```python
model = AgglomerativeClustering(n_clusters=None,distance_threshold=0)
cluster_labels = model.fit_predict(scaled_df)
cluster_labels
```

```python
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster import hierarchy
linkage_matrix = hierarchy.linkage(model.children_)
linkage_matrix # C0 is cluster 1, C1 is cluster 2, C2 is distance between them, C3 is the number of points in this cluster
```

```python
plt.figure(figsize=(20,10))
dendro = dendrogram(linkage_matrix)
```

```python
model = AgglomerativeClustering(n_clusters=None,distance_threshold=np.sqrt(len(scaled_df.columns)))
cluster_labels = model.fit_predict(scaled_df)
print(pd.Series(cluster_labels).value_counts())
linkage_matrix = hierarchy.linkage(model.children_)
plt.figure(figsize=(20,10))
dendro = dendrogram(linkage_matrix,truncate_mode='level',p=3)
```
