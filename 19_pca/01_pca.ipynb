```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('../course_notebooks/DATA/cancer_tumor_data_features.csv')
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_x = scaler.fit_transform(df)
```

```python
from sklearn.decomposition import PCA
pca_model = PCA(n_components=2)
```

```python
pca_model.fit(scaled_x)
```

```python
results = pca_model.transform(scaled_x)
results
```

```python
pca_model.explained_variance_
```

```python
pca_model.explained_variance_ratio_
```

```python
pca_model.components_
```

```python
sns.scatterplot(x=results[:,0],y=results[:,1])
```

```python
df_comp = pd.DataFrame(pca_model.components_,index=['PC1','PC2'],columns=df.columns)
df_comp
```

```python
np.sum(pca_model.explained_variance_ratio_)
```

```python
pca_30 = PCA(n_components=30)
pca_30.fit(scaled_x)
np.sum(pca_30.explained_variance_ratio_)
```

```python
explained_variance = []
for n in range(1,30):
    pca = PCA(n_components=n)
    pca.fit(scaled_x)
    explained_variance.append(np.sum(pca.explained_variance_ratio_))
plt.plot(range(1,30),explained_variance)
plt.xlabel('num of components')
plt.ylabel('total variance explained')
plt.ylim(0,1.0)
```
