```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
# .jpg ----> numpy array
```

```python
image_as_array = mpimg.imread('../course_notebooks/DATA/palm_trees.jpg')
```

```python
image_as_array.shape # (H,W,C) height, width, color channels
```

```python
plt.imshow(image_as_array)
```

```python
(h,w,c) = image_as_array.shape
```

```python
print(h)
print(w)
print(c)
```

```python
image_as_array2d = image_as_array.reshape(h*w,c) # flattening
```

```python
image_as_array2d
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=6)
labels = model.fit_predict(image_as_array2d)
```

```python
labels
```

```python
rgb_codes = model.cluster_centers_.round(0).astype(int)
rgb_codes
```

```python
rgb_codes[labels]
```

```python
np.reshape(rgb_codes[labels],(h,w,c))
```

```python
quantized_image = np.reshape(rgb_codes[labels],(h,w,c))
plt.imshow(quantized_image)
```

```python
model = KMeans(n_clusters=20)
labels = model.fit_predict(image_as_array2d)
rgb_codes = model.cluster_centers_.round(0).astype(int)
quantized_image = np.reshape(rgb_codes[labels],(h,w,c))
plt.imshow(quantized_image)
```
