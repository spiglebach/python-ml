```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../course_notebooks/08-Linear-Regression-Models/Advertising.csv')
df.head()
```

```python
df.corr()
```

```python
df['total_spend']=df['TV']+df['radio']+df['newspaper']
df.head()#total_spend is an engineered feature
```

```python
sns.scatterplot(data=df,x='total_spend',y='sales')
```

```python
sns.regplot(data=df,x='total_spend',y='sales')
```

```python
x=df['total_spend']
y=df['sales']
```

```python
# y=mx+b
# y=B1x + B0
help(np.polyfit)
```

```python
np.polyfit(x,y,deg=1)
```

```python
potential_advertisements = np.linspace(0,500,100)
predicted_sales = 0.048688 * potential_advertisements + 4.243028
```

```python
sns.scatterplot(data=df,x='total_spend',y='sales')
plt.plot(potential_advertisements,predicted_sales,color='red')
```
